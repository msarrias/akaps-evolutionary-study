### Phosphatase anchoring pairs

### PKA - RII subunit binding domain

```python
filenames = [file_name for file_name in os.listdir('fasta_files/') if '.fa' in file_name]
hmm_for_PKA_binding = binding_regions()
PKA_hmm_hits = {}
for filename in filenames:
    prot_ID = filename.rsplit('_oma')[0]
    hmm_for_PKA_binding.search_binding_regions("sto_files/"+ prot_ID +"_PKA.sto",
                                               "../../profile_hmm/hmm/PKA_binding.hmm",
                                               "fasta_files/" + filename)
    PKA_hmm_hits[prot_ID] = read_sto_files("sto_files/"+ prot_ID +"_PKA.sto")
```

```python
PKA_binding_prots = summary('RII subunit binding domain', PKA_hmm_hits, hmm_for_PKA_binding)
```

```python
missing_seqs, regions_dict, aln = align_ref_region(PKA_hmm_hits,
                                                    PKA_binding_prots,
                                                    hmm_for_PKA_binding,
                                                    'PKA_binding')
```

```python
visualize_aln(aln, missing_seqs)
```
