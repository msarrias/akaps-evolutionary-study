### Binding partners:

### 1. Calmodulin (CaM) - mediated by the WSK domain - [Patel et al. Nature (2017)](https://www.nature.com/articles/s41467-017-01715-w)

### I WSK instance

```python
# We will use as a reference the human architecture
ref = 'Homo_sapiens'
# calmodulin binding region - based on Patel et al.
base_string_wski = 'EGVTPWASFKKMVTPKKRVRRPSESDKE'
start, end = Gravin_model_species.find_binding_region(base_string_wski, ref, Gravin_msa)
start_m, end_m = Gravin_model_species.find_binding_region('WASFK', ref, Gravin_msa)
wski_residues_idx = (start_m,start_m+2,end_m-1)
filename = 'fasta_files/binding_regions/Gravin_WSK_i.fa'
regions_dict_wski, aln_wski = seq_domain_alignment(Gravin_msa,
                                                   Gravin_seqs,
                                                   start, end,
                                                   filename,
                                                   motif_coords=wski_residues_idx)
```

```python
print(f'The 1st WSK instance is missing in the following model species: {[i for i in Gravin_seqs.keys() if i not in regions_dict_wski.keys()]}')

alv.view(aln_wski)
```

#### Construct a hmm for identifying *calmodulin* binding regions in our model species

```python
new_seeds = {'homo_sapiens_gravin/605-633':base_string_wski,
             'homo_sapiens_gravin/754-782':base_string_wskii,
             'homo_sapiens_gravin/800-826':base_string_wskiii}
Gravin_model_species.generate_seed_file('profile_hmm/seed_files/PF03832_seed_v2.fa',
                                      new_seeds,
                                      'profile_hmm/seed_files/PF03832_seed_v3.fa')
```

```python
msa_filename = 'fasta_files/msa/PF03832_seed_v3_msa.fa'
# align the seed sequences - downloaded from Pfam
Gravin_model_species.muscle_msa('profile_hmm/seed_files/PF03832_seed_v3.fa', msa_filename)
# build the profile hmm
Gravin_model_species.build_profile_hmm("profile_hmm/hmm/WSK_v3.hmm", msa_filename)
# search for the binding region
Gravin_model_species.search_binding_regions("profile_hmm/WSK_v3.sto","profile_hmm/hmm/WSK_v3.hmm",seq_filename)
wsk_v3_hmm_hits = read_sto_files("profile_hmm/WSK_v3.sto")
```
