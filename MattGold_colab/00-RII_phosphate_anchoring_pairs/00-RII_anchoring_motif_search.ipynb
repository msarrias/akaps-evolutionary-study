### Phosphatase anchoring pairs

### PKA - RII subunit binding domain

```python
filenames = [file_name for file_name in os.listdir('fasta_files/') if '.fa' in file_name]
hmm_for_PKA_binding = binding_regions()
PKA_hmm_hits = {}
for filename in filenames:
    prot_ID = filename.rsplit('_oma')[0]
    hmm_for_PKA_binding.search_binding_regions("sto_files/"+ prot_ID +"_PKA.sto",
                                               "../../profile_hmm/hmm/PKA_binding.hmm",
                                               "fasta_files/" + filename)
    PKA_hmm_hits[prot_ID] = read_sto_files("sto_files/"+ prot_ID +"_PKA.sto")
```

```python
PKA_binding_prots = summary('RII subunit binding domain', PKA_hmm_hits, hmm_for_PKA_binding)
```

### PKA binding - mediated through the RII_binding_I domain

```python
hmm_for_RII_binding_I = binding_regions()
RII_binding_I_hmm_hits = {}
for filename in filenames:
    prot_ID = filename.rsplit('_oma')[0]
    hmm_for_RII_binding_I.search_binding_regions("sto_files/"+ prot_ID +"RII_binding.sto",
                                               "../../profile_hmm/hmm/RII_binding.hmm",
                                               "fasta_files/" + filename)
    RII_binding_I_hmm_hits[prot_ID] = read_sto_files("sto_files/"+ prot_ID +"RII_binding.sto")
```

```python
RII_binding_I_prots = summary('RII subunit binding domain', RII_binding_I_hmm_hits, hmm_for_RII_binding_I)
```

```python
ref_dict = {prot: [(key, value) for key,
                   value in RII_binding_I_hmm_hits[prot].items() if 'HUMAN' in key][0] 
            for prot in RII_binding_I_prots}
regions_dict = {}
aln = {}
missing_seqs = {}
for prot_ID in RII_binding_I_prots:
    msa = hmm_for_RII_binding_I.parse_fasta_file('fasta_files/msa/'+ prot_ID +'_oma_msa.fa')
    seqs = hmm_for_RII_binding_I.parse_fasta_file('fasta_files/'+ prot_ID +'_oma.fa')
    base_string = ref_dict[prot_ID][1][1]
    ref = ref_dict[prot_ID][0].rsplit('-')[0]
    start, end = hmm_for_RII_binding_I.find_binding_region(base_string, ref, msa)
    filename = 'fasta_files/binding_regions/'+ prot_ID +'RII_binding_I.fa'
    missing_seqs[prot_ID] = [id_ for id_ in seqs.keys() if id_ not in
       [i.rsplit('-')[0] for i in RII_binding_I_hmm_hits[prot_ID].keys()]]
    regions_dict[prot_ID], aln[prot_ID] = seq_domain_alignment(msa, seqs, start, end, filename,
                                                       binding_partner = 'RII_binding_I')
```
